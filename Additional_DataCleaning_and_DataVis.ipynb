## Preparing for Analysis (4 selected sites)

To view the PM 2.5 readings averaged by month across the slice, I have had to first split each slice by year- creating a new DataFrame for each year within that slice.  This avoids issues with groupby later on.

Once each year has been separated, monthly averages are taken using groupby and the DataFrames are concatenated together to create one DataFrame representing the monthly averages for that slice.

### PreCovid

```python
preCovid_2018 = preCovid[preCovid['Date_Time'].dt.year == 2018]
preCovid_monthly_2018 = preCovid_2018.groupby([preCovid_2018['Date_Time'].dt.month]).mean()

preCovid_2019 = preCovid[preCovid['Date_Time'].dt.year == 2019]
preCovid_monthly_2019 = preCovid_2019.groupby([preCovid_2019['Date_Time'].dt.month]).mean()

preCovid_2020 = preCovid[preCovid['Date_Time'].dt.year == 2020]
preCovid_monthly_2020 = preCovid_2020.groupby([preCovid_2020['Date_Time'].dt.month]).mean()
```

```python
preCovid_monthly = pd.concat([preCovid_monthly_2018, preCovid_monthly_2019, preCovid_monthly_2020])
```

I have added Year and Month columns for use in future analysis (makes using the groupby function easier)

```python
preCovid_monthly['Year'] = pd.DatetimeIndex(preCovid_monthly['Date_Time']).year
preCovid_monthly['Month'] = pd.DatetimeIndex(preCovid_monthly['Date_Time']).month
```

```python
preCovid_monthly.head(5)
```

```python
preCovid_months = list(preCovid_monthly['Month'])
preCovid_years = list(preCovid_monthly['Year'])

preCovid_labels = []

for i in range(len(preCovid_months)):
    l = (f"{preCovid_months[i]}-{str(preCovid_years[i])[2:]}")
    preCovid_labels.append(l)

#preCovid_labels
```

```python
sns.set(font_scale=5)
plt.figure(figsize = (80, 75))
sns.barplot(data=preCovid_monthly, x=preCovid_labels, y=preCovid_monthly['Salford M60'], hue='Year')
plt.ylabel("PM 2.5 (ug/m3)")
plt.xlabel('Month')
plt.title('Average PM 2.5 levels by month (Salford M60, PreCovid)')
```

```python
sns.set(font_scale=2)
sns.lineplot(data=preCovid_monthly, x='Month', y=preCovid_monthly['Salford M60'])
```

### Creating a column called Slice to indicate which original slice the data belongs to

This made it easier to use the hue argument when creating graphs)

```python
## create a column that indicates which slice that month falls into so that I can use hue/col = SLice
## i.e. if year == 2018, 2019 then preCovid, and if year == 2020 and month == jan/feb then precovid
## if year == 2020 and month is not jan/feb then covid, if year == 2021 then covid, if year == 2022 and month == jan then covid
## if year == 2022 and month is not jan then post covid, if year == 2023 or 2024 then post covid

four_2018 = four_sites[(four_sites['Year'] == 2018)]
four_2019 = four_sites[(four_sites['Year'] == 2019)]
four_2020 = four_sites[(four_sites['Year'] == 2020)]
four_2020_jan = four_2020[four_2020['Month'] == 1]
four_2020_feb = four_2020[four_2020['Month'] == 2]
four_2020_mar = four_2020[four_2020['Month'] == 3]

four_sites_precovid = pd.concat([four_2018, four_2019, four_2020_jan, four_2020_feb, four_2020_mar])
four_sites_precovid['Slice'] = 'pre-covid'

#four_sites_precovid
```
